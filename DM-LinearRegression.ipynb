```python
#Convert non-numeric columns to numeric whith LabelEncoder
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
HousePrice['Address'] = label_encoder.fit_transform(HousePrice['Address'])

#converting the bool type to int type
HousePrice["Parking"] = HousePrice["Parking"].astype(int)
HousePrice["Warehouse"] = HousePrice["Warehouse"].astype(int)
HousePrice["Elevator"] = HousePrice["Elevator"].astype(int)
HousePrice
```

```python
#Fill nulls whith KNNImputer
from sklearn.impute import KNNImputer
imputer = KNNImputer(n_neighbors=10, missing_values=np.nan)
df= imputer.fit_transform(HousePrice)
HousePrice=pd.DataFrame(df, index=HousePrice.index, columns=HousePrice.columns)
HousePrice
```

```python
#Check nulls
HousePrice.isnull().sum(axis=0)
```

```python
#Find outliers whith boxplot
sns.boxplot(data=HousePrice.loc[:, HousePrice.columns != "Price"])
```

```python
#Drop outliers whith z_score
from scipy import stats
HousePrice=HousePrice[np.abs(stats.zscore(HousePrice)<3).all(axis=1)]
HousePrice
```

```python
#Check outliers
import seaborn as sns
sns.boxplot(data=HousePrice.loc[:, HousePrice.columns != "Price"])
```

```python
#split x & y
x=HousePrice.drop(columns='Price')
y=HousePrice['Price']
```

```python
#normalization x
from sklearn.preprocessing import scale
scaled_x=scale(x)
x=pd.DataFrame(scaled_x,index=x.index, columns=x.columns)
x
```

```python
#train_test data split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
x_train
```

```python
#Fit LinearRegression model on dataset
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
reg.fit(x_train,y_train)
```

```python
#ُShow result and score of the model
from matplotlib import pyplot as plt
y_pred=reg.predict(x_test)
plt.scatter(y_test,y_pred)
plt.show()
reg.score(x_test,y_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
import statsmodels.api as sm
from scipy import stats

# model evaluation
mse=mean_squared_error(y_test,y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2 = r2_score(y_test, y_pred)

# printing values
print('Slope:' ,reg.coef_)
print('Intercept:', reg.intercept_)
print('MSE:',mse)
print('Root mean squared error: ', rmse)
print('R2 score: ', r2)
```
